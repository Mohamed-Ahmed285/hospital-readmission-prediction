```python

med_cols = [
    'metformin', 'repaglinide', 'nateglinide', 'glimepiride',
    'glipizide', 'glyburide', 'pioglitazone', 'rosiglitazone',
    'insulin', 'glyburide-metformin'
]


heatmap_data = pd.DataFrame()

for med in med_cols:
    
    grouped = data.groupby([med])['readmitted'].value_counts().unstack(fill_value=0)

    
    total = grouped.sum(axis=1)

    # Readmitted = sum of class 1 and 2
    readmit = grouped.get(1, 0) + grouped.get(2, 0)

    # Compute readmission rate
    heatmap_data[med] = (readmit / total).round(2)

# Optional: Convert index to string for better readability
heatmap_data.index = heatmap_data.index.astype(str)

# Plot
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data.T, annot=True, cmap="YlOrRd", cbar_kws={'label': 'Readmission Rate'})
plt.title("Readmission Rate by Medication Usage Level")
plt.xlabel("Usage Level")
plt.ylabel("Medication")
plt.tight_layout()
plt.show()
```
